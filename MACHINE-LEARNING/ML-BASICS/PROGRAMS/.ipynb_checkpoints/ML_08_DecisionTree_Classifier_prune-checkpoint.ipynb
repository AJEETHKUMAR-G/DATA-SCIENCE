## LIBRARY IMPORT

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
import numpy as np
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.model_selection import train_test_split
```

## DATA IMPORT

```python
cancer = load_breast_cancer()
```

```python
cancer
```

```python
df = pd.DataFrame(data = cancer.data,columns = cancer.feature_names)
```

```python
df
```

```python
df.shape
```

## SPLIT X AND Y

```python
x = df.copy()
y = cancer.target
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.33,random_state = 99)
```

## MODEL BUILDING

```python
clf = DecisionTreeClassifier(ccp_alpha = 0.01)
clf_dt = clf.fit(x_train,y_train)
```

```python
clf_dt
```

```python
clf.get_params()
```

```python
y_pred = clf_dt.predict(x_test)
```

```python
y_pred
```

```python
clf.predict_proba(x_test)
```

## EVALUATION METRICS

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```

## FEATURE IMPORTANCE

```python
feature_names = x.columns
feature_names
```

```python
feature_importance = pd.DataFrame(clf.feature_importances_,index = feature_names)
```

```python
feature_importance
```

```python
f = list(feature_importance[feature_importance[0]>0].index)
f
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (250,40))
plot_tree(clf_dt,filled = True,feature_names = feature_names,class_names = {0:'Mal',1:'Benign'})
plt.show()
```

## PRUNING

```python
x1,y1 = load_breast_cancer(return_X_y=True)
```

```python
x_train1,x_test1,y_train1,y_test1 = train_test_split(x1,y1,random_state = 0)
```

```python
clf = DecisionTreeClassifier(random_state = 0)
path = clf.cost_complexity_pruning_path(x_train1,y_train1)
```

```python
path
```

```python
ccp_alpha = path.ccp_alphas[:-1]
```

```python
ccp_alpha
```

```python
clfs = []
for i in ccp_alpha:
    clf = DecisionTreeClassifier(ccp_alpha = i,random_state = 0)
    clf.fit(x_train1,y_train1)
    clfs.append(clf)
```

```python
clfs
```

```python
train_scores = [clf.score(x_train1,y_train1) for clf in clfs]
```

```python
train_scores
```

```python
test_scores = [clf.score(x_test1,y_test1) for clf in clfs]
```

```python
test_scores
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for training and testing sets")
ax.plot(ccp_alpha, train_scores, marker="o", label="train", drawstyle="steps-post")
ax.plot(ccp_alpha, test_scores, marker="o", label="test", drawstyle="steps-post")
ax.legend()
plt.show()
```

```python
## CHOOSING IDEAL ALPHA = 0.015
```

```python
clf_prune = DecisionTreeClassifier(random_state = 0,ccp_alpha = 0.015)
cdt = clf_prune.fit(x_train1,y_train1)
```

```python
plt.figure(figsize = (250,40))
plot_tree(cdt,filled = True,feature_names = feature_names,class_names = {0:'Mal',1:'Benign'})
plt.show()
```
