## LIBRARY IMPORT

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import load_diabetes
from sklearn.tree import DecisionTreeRegressor,plot_tree
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.model_selection import train_test_split,cross_val_score
```

## DATA IMPORT

```python
dt = load_diabetes()
```

```python
x = pd.DataFrame(dt.data,columns = dt.feature_names)
y = dt.target
```

```python
x
```

```python
y
```

## TRAIN TEST SPLIT

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

## MODEL BUILDING

```python
reg = DecisionTreeRegressor(random_state = 42)
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
y_pred = reg.predict(x_test)
```

```python
y_test
```

```python
y_pred
```

```python
plt.figure(figsize = (15,8))
plot_tree(reg,filled = True,feature_names = x.columns)
plt.show()
```

## PRUNING

```python
path = reg.cost_complexity_pruning_path(x_train,y_train)
```

```python
path
```

```python
ccp_alpha = path.ccp_alphas
```

```python
ccp_alpha = ccp_alpha[:-1]
```

```python
ccp_alpha
```

```python
alpha_loop = []
for i in ccp_alpha:
    clf = DecisionTreeRegressor(random_state = 42,ccp_alpha = i)
    scores = cross_val_score(clf,x_train,y_train,cv = 5,scoring = 'r2')
    alpha_loop.append([i,np.mean(scores),np.std(scores)])
```

```python
alpha_res = pd.DataFrame(alpha_loop,columns = ['alpha','mean_r2','std_r2'])
alpha_res
```

```python
alpha_res.sort_values(by ='mean_r2',ascending = False)
```

```python
ideal = alpha_res.loc[319,'alpha']
```

```python
final_model = DecisionTreeRegressor(random_state = 42,ccp_alpha = ideal)
final_model.fit(x_train,y_train)
```

```python
y_pred_test = final_model.predict(x_test)
```

```python
r2_score(y_test,y_pred_test)
```

```python
mean_squared_error(y_test,y_pred_test)
```

```python
r2_score(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
plt.figure(figsize = (15,8))
plot_tree(final_model,filled = True,feature_names = x.columns)
plt.show()
```
